```python
data = [
    {'Name': "John",'Age':25,'gender':'M','salary':2000},
    {'Name': "Alice",'Age':30,'gender':'Female','salary':3000},
    {'Name': "Bob",'Age':None,'gender':'M','salary':2500},
    {'Name': "Pratik",'Age':22,'gender':'m','salary': None},
    {'Name': "John",'Age':25,'gender':'M','salary':2000},
    {'Name': "Babita",'Age':-10,'gender':'f','salary':4000},

]
```

```python
df= pd.DataFrame(data)
print(df)
```

```python
df.isnull().sum()
```

```python
# Replace negative age values with NaN
df['Age'] = df['Age'].apply(lambda x: np.nan if x < 0 else x)
```

```python
df['salary'] = df['salary'].fillna(df['salary'].mean())
# Fill NaN values with the mean of the 'Age' column
df['Age'] = df['Age'].fillna(df['Age'].mean())

df.isnull().sum()
```

```python
df.duplicated().sum()
print(df.duplicated())
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
print(df.duplicated())
```

```python
# Convert gender to lowercase
df['gender'] = df['gender'].str.lower()

# Map 'female' to 'f' for consistency
df['gender'] = df['gender'].replace({'female': 'f'})
# Map 'male' to 'm' for consistency
df['gender'] = df['gender'].replace({'male': 'm'})

print(df)
```

```python
#scaling
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['Age'] = scaler.fit_transform(df[['Age']])
df['salary'] = scaler.fit_transform(df[['salary']])

print(df)
```

```python
import pandas as pd
import numpy as np
df1 = pd.read_excel(r"/content/SY_Data.xlsx")
df1.head()
```

```python
df1.shape
```

```python
df1.describe()
```

```python
df1.isnull().sum()
```

```python
df1.duplicated().sum()
print(df1.duplicated())
```

```python
# Step 1: Convert to numeric, forcing errors to NaN
df1['Pattern'] = pd.to_numeric(df1['Pattern'], errors='coerce')

# Step 2 (Optional): Replace NaNs with 0 or some default value
df1['Pattern'] = df1['Pattern'].fillna(0)

# Step 3: Convert to int
df1['Pattern'] = df1['Pattern'].astype(int)
```

```python
# Step 1: Calculate median excluding 0 values
median_val = df1.loc[df1['Pattern'] != 0, 'Pattern'].median()

# Step 2: Replace 0 with the calculated median
df1['Pattern'] = df1['Pattern'].replace(0, median_val)

# Optional: If you want it as int
df1['Pattern'] = df1['Pattern'].astype(int)
```

```python
df1.isnull().sum()
```

```python
#scaling
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df1['Pattern'] = scaler.fit_transform(df1[['Pattern']])

print(df1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Bar plot
plt.figure(figsize=(10, 6))
sns.countplot(data=df1, x='Academic Year', hue='Course Code')
plt.title('Course Code Count per Academic Year')
plt.xlabel('Academic Year')
plt.ylabel('Number of Courses')
plt.legend(title='Course Code')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
gender_subject_counts = df1.groupby(['Gender', 'Course Code']).size().reset_index(name='Count')
print(gender_subject_counts.head())
```

```python
division_subject_counts = df1.groupby(['Division', 'Course Code']).size().reset_index(name='Count')
print(division_subject_counts.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.barplot(data=gender_subject_counts, x='Course Code', y='Count', hue='Gender')
plt.title('Subject-wise Count per Gender')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=division_subject_counts, x='Course Code', y='Count', hue='Division')
plt.title('Subject-wise Count per Division')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
